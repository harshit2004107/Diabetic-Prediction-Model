- Separating data and labels

```python
x = df.drop(columns='Outcome', axis = 1)
x.head()
```

```python
y = df['Outcome']
y.head()
```

- Data Standardization

```python
scalar = StandardScaler()
```

```python
scalar.fit(x)
```

```python
standardized_data = scalar.transform(x)
standardized_data
```

```python
X = standardized_data
Y = df['Outcome']
```

- train test split


```python
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.2, stratify=y, random_state = 2)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python

test_y.shape
```

- Model Training

- LOGISTIC REGRESSION


```python
lr = LogisticRegression()
lr = lr.fit(train_x, train_y)
lr
```

```python
logistic_train_pred = lr.predict(train_x)
logistic_test_pred = lr.predict(test_x)
# print(logistic_test_pred)
```

```python
print("Logistic Regression Training Accuracy: ", round(accuracy_score(train_y, logistic_train_pred), 2)*100)
print("Logistic Regression Testing Accuracy: ", round(accuracy_score(test_y, logistic_test_pred), 2)*100)
```

Insights: Logistic Regression is a simple and interpretable algorithm commonly used for binary classification tasks. It models the probability of the outcome using a logistic function, making it suitable for problems where the target variable is binary.
Reasons for Choosing: Logistic Regression is often the baseline algorithm for binary classification tasks due to its simplicity and interpretability. It's a good starting point for modeling and can provide insights into the relationship between the features and the target variable.

```python
from sklearn.metrics import roc_curve, auc

# Predict probabilities for training and test data
train_probs = lr.predict_proba(train_x)[:, 1]
test_probs = lr.predict_proba(test_x)[:, 1]

# Compute ROC curve for training data
fpr_train, tpr_train, thresholds_train = roc_curve(train_y, train_probs)
roc_auc_train = auc(fpr_train, tpr_train)

# Compute ROC curve for test data
fpr_test, tpr_test, thresholds_test = roc_curve(test_y, test_probs)
roc_auc_test = auc(fpr_test, tpr_test)

# Plot ROC curves
plt.figure(figsize=(10, 5))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (train) - AUC = {:.2f}'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, color='navy', lw=2, label='ROC curve (test) - AUC = {:.2f}'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

- Decision Trees

```python
max_accuracy = 0   
for i in range(2000):
    clf = DecisionTreeClassifier(random_state = i)
    clf.fit(train_x, train_y)
    pred = clf.predict(test_x)
    current_accuracy = round(accuracy_score(pred, test_y)*100,2)
    if(current_accuracy > max_accuracy):
        max_accuracy = current_accuracy
        best_random_state = i

print("Best Random State: ", best_random_state)
# print("Best Accuracy: ", max_accuracy)

clf = DecisionTreeClassifier(random_state = best_random_state)
clf.fit(train_x, train_y)
```

```python
plt.figure(figsize=(5,5))
tree.plot_tree(clf)
```

```python
decision_train_pred = clf.predict(train_x)
decision_test_pred = clf.predict(test_x)
```

```python
print("Decision Tree Training Accuracy: ", round(accuracy_score(train_y, decision_train_pred), 2)*100)
print("Decision Tree Testing Accuracy: ", round(accuracy_score(test_y, decision_test_pred), 2)*100)
```

Insights: Decision Trees are non-linear models that partition the feature space into regions and make predictions based on the majority class of the instances within each region. They are prone to overfitting, as evident from the perfect training accuracy and lower testing accuracy.
Reasons for Choosing: Decision Trees are useful for capturing complex non-linear relationships in the data. They are easy to interpret and visualize, making them suitable for gaining insights into how different features contribute to the decision-making process.

```python
from sklearn.metrics import roc_curve, auc

# Predict probabilities for training and test data
train_probs = clf.predict_proba(train_x)[:, 1]
test_probs = clf.predict_proba(test_x)[:, 1]

# Compute ROC curve for training data
fpr_train, tpr_train, thresholds_train = roc_curve(train_y, train_probs)
roc_auc_train = auc(fpr_train, tpr_train)

# Compute ROC curve for test data
fpr_test, tpr_test, thresholds_test = roc_curve(test_y, test_probs)
roc_auc_test = auc(fpr_test, tpr_test)

# Plot ROC curves
plt.figure(figsize=(10, 5))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (train) - AUC = {:.2f}'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, color='navy', lw=2, label='ROC curve (test) - AUC = {:.2f}'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

- SVM

```python
classifier = SVC(kernel = 'linear')
classifier.fit(train_x, train_y)
```

```python
svm_test_pred = classifier.predict(test_x)
```

```python
print("SVM Training Accuracy: ", round(classifier.score(train_x, train_y)*100, 2))
print("SVM Testing Accuracy: ", round(classifier.score(test_x, test_y)*100, 2))
```

Insights: SVMs are powerful classifiers that find the optimal hyperplane separating different classes in the feature space. They are effective in high-dimensional spaces and can handle non-linear decision boundaries through the use of different kernel functions.
Reasons for Choosing: SVMs are suitable for scenarios where the decision boundary between classes is complex or non-linear. They work well with both linear and non-linear data and offer good generalization performance. SVMs are chosen here for their ability to handle complex classification tasks and their high accuracy on the testing set.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Predict decision function scores for training and test data
train_scores = classifier.decision_function(train_x)
test_scores = classifier.decision_function(test_x)

# Convert decision scores to probabilities using Platt scaling
train_probs = 1 / (1 + np.exp(-train_scores))
test_probs = 1 / (1 + np.exp(-test_scores))

# Compute ROC curve for training data
fpr_train, tpr_train, thresholds_train = roc_curve(train_y, train_probs)
roc_auc_train = auc(fpr_train, tpr_train)

# Compute ROC curve for test data
fpr_test, tpr_test, thresholds_test = roc_curve(test_y, test_probs)
roc_auc_test = auc(fpr_test, tpr_test)

# Plot ROC curves
plt.figure(figsize=(10, 5))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (train) - AUC = {:.2f}'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, color='navy', lw=2, label='ROC curve (test) - AUC = {:.2f}'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
